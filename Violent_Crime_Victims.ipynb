# Victims of Violent Crimes by Gender in Hamilton

```python
# Import dependencies
import matplotlib.pyplot as plt
import pandas as pd
from datetime import date
from pathlib import Path
today = date.today()
```

```python
# File to load
file_path = Path("Resources/Violent Crime by Sex of Victim - Hamilton, Ontario and Canada.xlsx")

# Read the excel file into a dataframe
victim_df = pd.read_excel(file_path)

# Display the data
victim_df.head()
```

```python
# Check data types
victim_df.dtypes
```

```python
# Convert 'Year' column to datetime
victim_df['Year'] = pd.to_datetime(victim_df['Year'], format='%Y')
```

```python
# Check data types
victim_df.dtypes
```

```python
# Set the index to the 'Year' Column
victim_df = victim_df.set_index('Year')

# Display the date
victim_df.head()
```

```python
# Lines for the plot
plt.plot(victim_df.index, victim_df['Female - Hamilton'])
plt.plot(victim_df.index, victim_df['Male - Hamilton'])

# Title of the plot
plt.title('Number of Victims of Violent Crimes by Gender in Hamilton (2009-2018)')
# Bold title
plt.title('Number of Victims of Violent Crimes by Gender in Hamilton (2009-2018)', fontweight='bold')

# Label x and y axis
plt.xlabel('Year')
plt.ylabel('Number of Victims')

# Add legend
plt.legend(['Female', 'Male'])

# Plot the data
plt.show()
```

```python
# Lines for the plot
plt.plot(victim_df.index, victim_df['Female - Ontario'])
plt.plot(victim_df.index, victim_df['Male - Ontario'])

# Title of the plot
plt.title('Number of Victims of Violent Crimes by Gender in Ontario (2009-2018)')
# Bold title
plt.title('Number of Victims of Violent Crimes by Gender in Ontario (2009-2018)', fontweight='bold')

# Label x and y axis
plt.xlabel('Year')
plt.ylabel('Number of Victims')

# Add legend
plt.legend(['Female', 'Male'])

# Plot the data
plt.show()
```

```python
# Lines for the plot
plt.plot(victim_df.index, victim_df['Female - Canada'])
plt.plot(victim_df.index, victim_df['Male - Canada'])

# Title of the plot
plt.title('Number of Victims of Violent Crimes by Gender in Canada (2009-2018)')
# Bold title
plt.title('Number of Victims of Violent Crimes by Gender in Canada (2009-2018)', fontweight='bold')

# Label x and y axis
plt.xlabel('Year')
plt.ylabel('Number of Victims')

# Add legend
plt.legend(['Female', 'Male'])

# Plot the data
plt.show()
```

```python
# Lines for the plot
plt.plot(victim_df.index, victim_df['Female - Hamilton'])
plt.plot(victim_df.index, victim_df['Male - Hamilton'])
plt.plot(victim_df.index, victim_df['Female - Ontario'], linestyle='dashed')
plt.plot(victim_df.index, victim_df['Male - Ontario'], linestyle='dashed')
plt.plot(victim_df.index, victim_df['Female - Canada'], linestyle='dotted')
plt.plot(victim_df.index, victim_df['Male - Canada'], linestyle='dotted')

# Title of the plot
plt.title('Number of Victims of Violent Crimes by Gender (2009-2018)')
# Bold title
plt.title('Number of Victims of Violent Crimes by Gender (2009-2018)', fontweight='bold')

# Label x and y axis
plt.xlabel('Year')
plt.ylabel('Number of Victims')

# Add legend
plt.legend(['Female - Hamilton', 'Male- Hamilton', 'Female - Ontario', 
            'Male - Ontario', 'Female - Canada', 'Male - Canada'])

# Plot the data
plt.show()
```
