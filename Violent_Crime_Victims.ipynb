# Victims of Violent Crimes by Gender in Hamilton, Ontario and Canada

```python
# Import dependencies
import matplotlib.pyplot as plt
import pandas as pd
from datetime import date
from pathlib import Path
today = date.today()
```

```python
# File to load
file_path = Path("Resources/Violent Crime by Sex of Victim - Hamilton, Ontario and Canada.xlsx")

# Read the excel file into a dataframe
victim_df = pd.read_excel(file_path)

# Display the data
victim_df.head()
```

```python
# Check data types
victim_df.dtypes
```

```python
# Convert 'Year' column to datetime
victim_df['Year'] = pd.to_datetime(victim_df['Year'], format='%Y')
```

```python
# Check data types
victim_df.dtypes
```

```python
# Set the index to the 'Year' Column
victim_df = victim_df.set_index('Year')

# Display the date
victim_df.head()
```

```python
# Summary Stats

summary_stats = victim_df.describe()
print(summary_stats)
```

```python
# Correlation
victim_df.corr()
```

```python
# Lines for the plot
plt.plot(victim_df.index, victim_df['Female - Hamilton'], color='pink',
        linewidth=2, marker='o', markersize=4, label='Female')
plt.plot(victim_df.index, victim_df['Male - Hamilton'], color='blue',
        linewidth=2, marker='o', markersize=4, label='Male')

# Title of the plot
plt.title('Number of Victims of Violent Crimes by Gender in Hamilton Between 2009-2018', fontweight='bold')


# Label x and y axis
plt.xlabel('Year')
plt.ylabel('Number of Victims')

# Add legend
plt.legend()

# Save the plot
plt.savefig('Images/Number of Victims of Violent Crimes by Gender in Hamilton Between 2009-2018.png', bbox_inches='tight')

# Plot the data
plt.show()
```

```python
# Lines for the plot
plt.plot(victim_df.index, victim_df['Female - Ontario'], color='pink',
        linewidth=2, marker='o', markersize=4, label='Female')
plt.plot(victim_df.index, victim_df['Male - Ontario'], color='blue',
        linewidth=2, marker='o', markersize=4, label='Male')

# Title of the plot
plt.title('Number of Victims of Violent Crimes by Gender in Ontario Between 2009-2018', fontweight='bold')

# Label x and y axis
plt.xlabel('Year')
plt.ylabel('Number of Victims')

# Add legend
plt.legend()

# Save the plot
plt.savefig('Images/Number of Victims of Violent Crimes by Gender in Ontario Between 2009-2018.png', bbox_inches='tight')

# Plot the data
plt.show()
```

```python
# Lines for the plot
plt.plot(victim_df.index, victim_df['Female - Canada'], color='pink',
        linewidth=2, marker='o', markersize=4, label='Female')
plt.plot(victim_df.index, victim_df['Male - Canada'], color='blue',
        linewidth=2, marker='o', markersize=4, label='Male')

# Title of the plot
plt.title('Number of Victims of Violent Crimes by Gender in Canada Between 2009-2018)', fontweight='bold')

# Label x and y axis
plt.xlabel('Year')
plt.ylabel('Number of Victims')

# Add legend
plt.legend()

# Save the plot
plt.savefig('Images/Number of Victims of Violent Crimes by Gender in Canada Between 2009-2018.png', bbox_inches='tight')

# Plot the data
plt.show()
```

```python
# Lines for the plot
plt.plot(victim_df.index, victim_df['Female - Hamilton'], color='pink',
        linewidth=2, marker='o', markersize=4, label='Female - Hamilton')
plt.plot(victim_df.index, victim_df['Male - Hamilton'], color='blue',
        linewidth=2, marker='o', markersize=4, label='Male - Hamilton')
plt.plot(victim_df.index, victim_df['Female - Ontario'], linestyle='dashed',
        color='hotpink',linewidth=2, marker='o', markersize=4, 
         label='Female - Ontario')
plt.plot(victim_df.index, victim_df['Male - Ontario'], linestyle='dashed',
        color='steelblue', linewidth=2, marker='o', markersize=4, 
         label='Male - Ontario')
plt.plot(victim_df.index, victim_df['Female - Canada'], linestyle='dotted',
        color='mediumvioletred',linewidth=2, marker='o', markersize=4, 
         label='Female - Canada')
plt.plot(victim_df.index, victim_df['Male - Canada'], linestyle='dotted',
        color='skyblue',linewidth=2, marker='o', markersize=4, 
         label='Male - Canada')

# Title of the plot
plt.title('Number of Victims of Violent Crimes by Gender Between 2009-2018', fontweight='bold')

# Label x and y axis
plt.xlabel('Year')
plt.ylabel('Number of Victims')

# Add legend
plt.legend()

# Save the plot
plt.savefig('Images/Number of Victims of Violent Crimes by Gender Between 2009-2018.png', bbox_inches='tight')

# Plot the data
plt.show()
```
