# Hate Crimes as Reported by Hamilton Police Services 2019-2023

```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import date
from pathlib import Path
today = date.today()
```

```python
# Load file
data_path = Path("Resources/Hate Crimes - HPS - 2019-2023.xlsx")

# Read in the data
hate_crimes = pd.read_excel(data_path)

# Display the data
hate_crimes.head()
```

```python
# Create DataFrame
hate_crimes_df = pd.DataFrame(hate_crimes)

# Display the data
hate_crimes_df.head()
```

```python
# Check data types
hate_crimes_df.dtypes
```

```python
# Set index to 'Date' column
hate_crimes_df = hate_crimes_df.set_index('Date')

# Display the data
hate_crimes_df.head()
```

```python
# Lines for the plot
plt.plot(hate_crimes_df.index, hate_crimes_df['Total - Bias'])
plt.plot(hate_crimes_df.index, hate_crimes_df['Total - Motivated '])
plt.plot(hate_crimes_df.index, hate_crimes_df['Overall Total'])

# Title of the plot
plt.title('Total Hate Crimes in Hamilton (2019-2023)')
# Bold title
plt.title('Total Hate Crimes in Hamilton (2019-2023)', fontweight='bold')

# Labels for the x and y axis
plt.xlabel('Date')
plt.ylabel('Number of Crimes')

# Place x ticks
plt.xticks(rotation=30, fontsize=10)

# Legend
plt.legend(['Total - Bias', 'Total - Motivated ', 'Overall Total'])

# Plot the data
plt.show()

# Save the plot
#plt.savefig(f"Images/Hate_Crimes_Hamilton_2019-2023.png")
```

```python
# Lines for the plot
plt.plot(hate_crimes_df.index, hate_crimes_df['Gender Identity - Bias'], linestyle='dashed', color='blue')
plt.plot(hate_crimes_df.index, hate_crimes_df['Gender Identity - Motivated '], color='blue')
plt.plot(hate_crimes_df.index, hate_crimes_df['Sex - Bias'], linestyle='dashed', color='green')
plt.plot(hate_crimes_df.index, hate_crimes_df['Sex - Motivated'], color='green')
plt.plot(hate_crimes_df.index, hate_crimes_df['Sexual Orientation - Bias '], linestyle='dashed', color='orange')
plt.plot(hate_crimes_df.index, hate_crimes_df['Sexual Orientation - Motivated'], color='orange')

# Title of the plot
plt.title('Hate Crimes in Hamilton (2019-2023)')
# Bold title
plt.title('Hate Crimes in Hamilton (2019-2023)', fontweight='bold')

# Labels for the x and y axis
plt.xlabel('Date')
plt.ylabel('Number of Crimes')

# Place x ticks
plt.xticks(rotation=30, fontsize=10)

# Legend
plt.legend(['Gender Identity - Bias', 'Gender Identity - Motivated ', 'Sex - Bias', 'Sex - Motivated', 
            'Sexual Orientation - Bias ', 'Sexual Orientation - Motivated'])

# Plot the data
plt.show()

# Save the plot
#plt.savefig(f"Images/Hate_Crimes_Hamilton_2019-2023.png")
```
