# Hate Crimes as Reported by Hamilton Police Services 2019-2023

```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import date
from pathlib import Path
today = date.today()
```

```python
# Load file
data_path = Path("Resources/Hate Crimes - HPS - 2019-2023.xlsx")

# Read in the data
hate_crimes = pd.read_excel(data_path)

# Display the data
hate_crimes.head()
```

```python
# Create DataFrame
hate_crimes_df = pd.DataFrame(hate_crimes)

# Display the data
hate_crimes_df.head()
```

```python
# Check data types
hate_crimes_df.dtypes
```

```python
# Set index to 'Date' column
hate_crimes_df = hate_crimes_df.set_index('Date')

# Display the data
hate_crimes_df.head()
```

```python
# Summary Stats

summary_stats = hate_crimes_df.describe()
print(summary_stats)
```

```python
# Correlation
hate_crimes_df.corr()
```

```python
# Lines for the plot
plt.plot(hate_crimes_df.index, hate_crimes_df['Total - Bias'], color='blue',
        linewidth=2, marker='o', markersize=4, label='Bias')
plt.plot(hate_crimes_df.index, hate_crimes_df['Total - Motivated '], color='orange',
        linewidth=2, marker='o', markersize=4, label='Motivated')
plt.plot(hate_crimes_df.index, hate_crimes_df['Overall Total'], color='green', 
        linewidth=2, marker='o', markersize=4, label='Combined Total')

# Title of the plot
plt.title('Hate Crimes in Hamilton Between 2019-2023', fontweight='bold')

# Labels for the x and y axis
plt.xlabel('Date')
plt.ylabel('Number of Crimes')

# Place x ticks
ticks = ['2019-12-31', '2020-12-31', '2021-12-31', '2022-12-31', '2023-12-31']
plt.xticks(ticks)

# Legend
plt.legend()

# Save the plot
plt.savefig('Images/Hate Crimes in Hamilton Between 2019-2023.png', bbox_inches='tight')

# Plot the data
plt.show()

```

```python
# Lines for the plot
plt.plot(hate_crimes_df.index, hate_crimes_df['Gender Identity - Bias'], linestyle='dashed', color='blue',
        linewidth=1, marker='o', markersize=4, label='Gender Identity - Bias')
plt.plot(hate_crimes_df.index, hate_crimes_df['Gender Identity - Motivated '], color='blue',
        linewidth=1, marker='o', markersize=4, label='Gender Identity - Motivated')
plt.plot(hate_crimes_df.index, hate_crimes_df['Sex - Bias'], linestyle='dashed', color='green', 
        linewidth=1, marker='o', markersize=4, label='Sex - Bias')
plt.plot(hate_crimes_df.index, hate_crimes_df['Sex - Motivated'], color='green',
        linewidth=1, marker='o', markersize=4, label='Sex - Motivated')
plt.plot(hate_crimes_df.index, hate_crimes_df['Sexual Orientation - Bias '], linestyle='dashed', color='orange',
        linewidth=1, marker='o', markersize=4, label='Sexual Orientation - Bias')
plt.plot(hate_crimes_df.index, hate_crimes_df['Sexual Orientation - Motivated'], color='orange', 
        linewidth=1, marker='o', markersize=4, label='Sexual Orientation - Motivated')

# Title of the plot
plt.title('Hate Crimes by Type in Hamilton Between 2019-2023', fontweight='bold')

# Labels for the x and y axis
plt.xlabel('Date')
plt.ylabel('Number of Crimes')

# Place x ticks
ticks = ['2019-12-31', '2020-12-31', '2021-12-31', '2022-12-31', '2023-12-31']
plt.xticks(ticks)

# Legend
plt.legend()

# Save the plot
plt.savefig('Images/Hate Crimes by Type in Hamilton Between 2019-2023.png', bbox_inches='tight')

# Plot the data
plt.show()
```
